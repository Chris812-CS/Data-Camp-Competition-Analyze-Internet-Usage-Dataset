## 2.2 Imputation Strategy

```r
# Remove 2023 column
data_clean <- internet_usage %>%
  select(-`2023`)

# Apply foward fill
data_clean <- data_clean %>%
  fill(`2000`:`2022`, .direction = "down")

# Apply backward fill for any remaining missing values
data_clean <- data_clean %>%
  fill(`2000`:`2022`, .direction = "up")

# Check for missing values remaining
plot_missing(data_clean)
```

To handle missing values, here's the steps to achieve:

1. Removing the 2023 Column: Since the 2023 column had an overwhelming percentage of missing data (72.81%), its removal was a practical choice to prevent skewed results or unreliable imputations.
2. Forward Fill (Filling Missing Data from Previous Values): This step ensured that any gaps in the data were filled using the most recent available value, which is a logical approach when dealing with time-series data where trends are assumed to be continuous.
3. Backward Fill (Filling Remaining Gaps Using Future Values): For any gaps that forward fill couldn't handle (e.g., if missing data appeared at the start of the dataset), backward fill ensured no gaps remained by propagating the next available value.
4. Final Result (Plot): The updated plot confirms that your preprocessing steps successfully addressed all missing data, leaving the dataset complete and ready for analysis.

## 2.3 Data Transformation

```r
# Transform datatypes to suitable ones
# Convert "Country Code" and "Country Name" to factors if they are categorical
data_clean <- data_clean %>%
  mutate(`Country Code` = as.factor(`Country Code`),
         `Country Name` = as.factor(`Country Name`))

# Convert year columns (2000 to 2022) to numeric to ensure calculations are possible
data_clean <- data_clean %>%
  mutate(across(`2000`:`2022`, as.numeric))

# 3. Show the final datatypes of the dataset
str(data_clean)
```
